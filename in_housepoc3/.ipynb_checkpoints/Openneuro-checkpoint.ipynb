```python
import os
import os.path as op
import openneuro

import mne
from mne.datasets import sample
from mne_bids import (
    BIDSPath,
    read_raw_bids,
    print_dir_tree,
    make_report,
    find_matching_paths,
    get_entity_vals,
)
```

```python
dataset = "ds002778"
subject = "pd6"

bids_root = op.join(op.dirname(sample.data_path()), dataset)
if not op.isdir(bids_root):
    os.makedirs(bids_root)

openneuro.download(dataset=dataset, target_dir=bids_root)
```

```python
print_dir_tree(bids_root, max_depth=4)
```

```python
sessions = get_entity_vals(bids_root, "session", ignore_sessions="on")
datatype = "eeg"
extensions = [".bdf", ".tsv"]
bids_paths = find_matching_paths(
    bids_root, datatypes=datatype, sessions=sessions, extensions=extensions
)
```

```python
print(bids_paths)
```

```python
session = "off"
bids_path = BIDSPath(root=bids_root, session=session, datatype=datatype)
print(bids_path.match(ignore_json=True))
```

```python
len(bids_path.match(ignore_json=True))
```

```python
task = "rest"
suffix = "eeg"

sub = ["pd11", "pd12", "pd13", "pd14", "pd16", "pd17", "pd19", "pd22", "pd23", "pd26", "pd28", "pd3", "pd5", "pd6", "pd9"]
data = []

for s in sub:
    bids_path = bids_path.update(subject=s, task=task, suffix=suffix)
    raw = read_raw_bids(bids_path=bids_path, verbose=False)
    events = mne.find_events(raw)
    data += [[raw, events]]
    #print(data)
    
senior = []
notsenior = []

for d in data :
    if int(d[0].info["subject_info"]["age"]) < 60:
        notsenior += [d]
    else:
        senior += [d]
        
print(len(senior))
print(len(notsenior))
        
```

```python
chs = ['Fp1', 'AF3', 'F7', 'F3', 'FC1', 'FC5', 'T7', 'C3', 'CP1', 'CP5', 'P7', 'P3', 'Pz', 'PO3', 'O1', 'Oz', 'O2', 'PO4', 'P4', 'P8', 'CP6', 'CP2', 'C4', 'T8', 'FC6', 'FC2', 'F4', 'F8', 'AF4', 'Fp2', 'Fz', 'Cz', 'EXG1', 'EXG2', 'EXG3', 'EXG4', 'EXG5', 'EXG6', 'EXG7', 'EXG8', 'Status']
```

```python
chan_idxs = [senior[0].ch_names.index(ch) for ch in chs]
senior[0].plot(order=chan_idxs, start=12, duration=4)
notsenior[0].plot(order=chan_idxs, start=12, duration=4)
```

```python
ep_senior = []
for s in senior:
    ep_senior += [mne.Epochs(s[0], s[1], tmin=-0.1, tmax=1, baseline=(None, 0), preload=True)]
ep_notsenior = []
for ns in notsenior:
    ep_notsenior += [mne.Epochs(ns[0], ns[1], tmin=-0.1, tmax=1, baseline=(None, 0), preload=True)]
```

```python
ep_senior[0].plot()
ep_notsenior[1].plot()
print(ep_senior[0].average())
```

```python
ev_senior = []
for es in ep_senior : 
    ev_senior += [es.average()]
    
ev_notsenior = []
for ens in ep_notsenior : 
    ev_notsenior += [ens.average()]
```

```python
for ns in ev_notsenior :
    ns.plot()
```

```python
av_senior = mne.grand_average(ev_senior)
av_notsenior = mne.grand_average(ev_notsenior[1:])
```

```python
av_senior.plot()
av_notsenior.plot()
```
